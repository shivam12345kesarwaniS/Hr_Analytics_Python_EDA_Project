```python


plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Age', kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='JobRole', y='MonthlyIncome', data=df)
plt.title('Job Role vs. Monthly Income')
plt.xlabel('Job Role')
plt.ylabel('Monthly Income')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15, 10))

# Total Working Years histogram
plt.subplot(221)
plt.hist(df['TotalWorkingYears'], bins=20, color='skyblue')
plt.title('Total Working Years Distribution')
plt.xlabel('Total Working Years')
plt.ylabel('Frequency')

# Education Level histogram
plt.subplot(222)
plt.hist(df['Education'], bins=5, color='salmon')
plt.title('Education Level Distribution')
plt.xlabel('Education Level')
plt.ylabel('Frequency')

# Number of Companies Worked histogram
plt.subplot(223)
plt.hist(df['NumCompaniesWorked'], bins=15, color='lightgreen')
plt.title('Number of Companies Worked Distribution')
plt.xlabel('Number of Companies Worked')
plt.ylabel('Frequency')

# Distance from Home histogram
plt.subplot(224)
plt.hist(df['DistanceFromHome'], bins=20, color='orange')
plt.title('Distance from Home Distribution')
plt.xlabel('Distance from Home')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
